# Step 4 Follow the exercise instructions for creating loyalty flag for existing customers using 'transform() and loc() functions.

```python
# Aggregating data with 'transform()' function

ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Deriving columns with 'loc()' function
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# Step 6 The team wants to target different types of spenders in their marketing.  This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following:

# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a "Low spender"

# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a "High spender"

```python
# Aggregating data with 'transform()' function

ords_prods_merge['type_of_spender'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# Deriving columns with 'loc()' function
```

```python
ords_prods_merge.loc[ords_prods_merge['type_of_spender'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['type_of_spender'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'spending_flag', 'prices']].head(60)
```

# Step 7 To send relevant notifications to users within the app, if they want to buy the same item again, the team wants 'frequent' vs 'non-frequent' customers. Create an 'order_frequency_flag' that marks the regularity of a user's ordering behavior according to the median in the "days_since_prior_order" column. The criteria for the flag is as follows:

# If the median of "days_since_prior_order" is higher than 20 then the customer should be labeled a "Non-frequent customer"

# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a "Regular customer"

# If the median is lower than or equal to 10, then the customer should be labeled a "Frequent customer"

```python
# Aggregating data with 'transform()' function

ords_prods_merge['regularity_of_user'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
# Deriving columns with 'loc()' function
```

```python
ords_prods_merge.loc[ords_prods_merge['regularity_of_user'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['regularity_of_user'] <= 20) & (ords_prods_merge['regularity_of_user'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['regularity_of_user'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'regularity_of_user', 'days_since_prior_order']].head(60)
```
